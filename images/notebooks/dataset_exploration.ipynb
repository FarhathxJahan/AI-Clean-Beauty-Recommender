```python
skincare_df = df[df["primary_category"] == "Skincare"]
```

```python
skincare_df.shape
```

```python
important_columns = [
    "product_name",
    "brand_name",
    "ingredients",
    "rating",
    "reviews",
    "price_usd",
    "highlights",
    "primary_category",
    "secondary_category"
]

skincare_df = skincare_df[important_columns]
```

```python
skincare_df.isnull().sum()
```

```python
skincare_df = skincare_df.dropna(subset=["ingredients"])
```

```python
skincare_df.shape
```

```python
skincare_df["rating"] = skincare_df["rating"].fillna(skincare_df["rating"].mean())
```

```python
skincare_df.to_csv("cleaned_skincare.csv", index=False)
```
