```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_skincare.csv")
```

```python
df.head()
```

```python
parabens = [
    "methylparaben",
    "propylparaben",
    "butylparaben",
    "ethylparaben",
    "isobutylparaben",
    "isopropylparaben"
]
sulfates = [
    "sodium lauryl sulfate",
    "sodium laureth sulfate",
    "ammonium lauryl sulfate"
]
drying_alcohols = [
    "alcohol denat",
    "ethanol",
    "isopropyl alcohol",
    "sd alcohol"
]
fragrance_allergens = [
    "fragrance",
    "parfum",
    "limonene",
    "linalool",
    "citral",
    "geraniol",
    "eugenol"
]
comedogenic_ingredients = [
    "coconut oil",
    "isopropyl myristate",
    "algae extract"
]
strong_actives = [
    "benzoyl peroxide",
    "high concentration aha",
    "high concentration bha"
]
silicones = [
    "dimethicone",
    "cyclopentasiloxane"
]
essential_oils = [
    "peppermint oil",
    "lavender oil",
    "eucalyptus oil",
    "citrus oil"
]
```

```python
potential_irritants = (
    parabens +
    sulfates +
    drying_alcohols +
    fragrance_allergens +
    comedogenic_ingredients +
    strong_actives +
    silicones +
    essential_oils
)
```

```python
df["ingredients"] = df["ingredients"].fillna("")
```

```python
df["ingredients"] = df["ingredients"].str.lower()
```

```python
def find_irritants(ingredient_text):

    found = []

    for ingredient in potential_irritants:

        if ingredient in ingredient_text:
            found.append(ingredient)

    return found
```

```python
df["detected_irritants"] = df["ingredients"].apply(find_irritants)
```

```python
df["irritant_count"] = df["detected_irritants"].apply(len)
```

```python
df["clean_score"] = 100 - (df["irritant_count"] * 5)
```

```python
df["clean_score"] = df["clean_score"].clip(lower=0)
```
